**Ejemplo 1 EOQ** 

La demanda semanal de un producto es de 1000 unidades y los artículos se retiran a una tasa constante. El costo de colocar una orden para reabastecer el inventario es de 40 dólares. El costo unitario por artículo es de 5 dólares y el costo de mantener un inventario es de $0.10 por artículo por semana.


a) Suponga que no se permiten faltantes. Determine cuándo y cuánto
debe ordenarse.

$Q^{*}=\sqrt{\frac{2AD}{h}}=\sqrt{\frac{2*40*1000}{0.10}} =894.427$

$T^{*}=\frac{Q^{*}}{D}=\sqrt{\frac{2A}{Dh}}=\frac{894.427}{1000}=0.8944$

b) Si se permiten faltantes semanales de 3 dólares por artículo, determine
cuándo y cuánto debe ordenarse.

$Q^{*}=\sqrt{\frac{2DA}{h}}\sqrt{\frac{p+h}{p}}=\sqrt{\frac{2*40*1000}{0.10}}\sqrt{\frac{3+0.10}{3}}=909.21$





```julia
#La informacion corresponde a 
h=0.10 #costo de mantener el inventario 
A=40 #costo de colocar un pedido
c=5
D=1000
p=3
# ----------------------
#punto A
Q=sqrt((2*A*D)/h)
println("El valor de Q* es=",Q)
T=Q/1000
println("El valor de T* es=",T)
#punto B
Q2=sqrt((2*A*D)/h)*sqrt((p+h)/p)
println("El valor de Q* con faltantes es=",Q2)
using Plots,DataFrames
pyplot()  
x=[0;Q/D;	Q/D;	2Q/D;	2Q/D]
y=[Q;0;	Q	;0;	Q]
y2=[Q;Q;Q;Q;Q]
plot(x,y,label="Q", line=(3),marker=4,xlabel="Tiempo",ylabel="Inventario")
plot!(x,y2,label="Q*",line=(2),line=(:dash,4))
```

**Ejemplo 1 EOQ** 
_________________________


La demanda semanal de un producto es de 1000 unidades y los artículos se retiran a una tasa constante. El costo de colocar una orden para reabastecer el inventario es de 40 dólares. El costo unitario por artículo es de 5 dólares y el costo de mantener un inventario es de $0.10 por artículo por semana.


a) Suponga que no se permiten faltantes. Determine cuándo y cuánto
debe ordenarse.

$Q^{*}=\sqrt{\frac{2AD}{h}}=\sqrt{\frac{2*40*1000}{0.10}} =894.427$

$T^{*}=\frac{Q^{*}}{D}=\sqrt{\frac{2A}{Dh}}=\frac{894.427}{1000}=0.8944$

b) Si se permiten faltantes semanales de 3 dólares por artículo, determine
cuándo y cuánto debe ordenarse.

$Q^{*}=\sqrt{\frac{2DA}{h}}\sqrt{\frac{p+h}{p}}=\sqrt{\frac{2*40*1000}{0.10}}\sqrt{\frac{3+0.10}{3}}=909.21$

**Nota**
<div class=text-justify>
Para propositos de modelado, Harris repesento tanto tiempo como productos como cantidades continuas. Desde que asumio constantes, demanda deterministica, poner ordenes para Q unidades cada vez el inventario alcanza cero da en promedio un inventario de Q/2. El costo de mantener inventario es entonces hQ/2 por periodo de tiempo. El costo de alistamiento de pedido AD/Q por oden por tiempo. El costo de material c es entonces cD por periodo   
</div>

$K(Q)^{*}=cD+\sqrt{2ADh}$ =5089.44

$Y(Q)=cD+\frac{hQ}{2}+\frac{AD}{Q}$=5089.44


```julia
#La informacion corresponde a 
h=0.10 #costo de mantener el inventario 
A=40 #costo de colocar un pedido
c=5
D=1000
p=3
# ----------------------
#punto A
Q=sqrt((2*A*D)/h)
println("El valor de Q* es=",Q)
T=Q/1000
println("El valor de T* es=",T)
YK=(Q*h/2)+(A*D/Q)+(c*D)
println("El valor de Y(Q)* es=",YK)
K2=c*D+sqrt(2A*D*h)
println("El valor de K(Q)* es=",K2)
#punto B
Q2=sqrt((2*A*D)/h)*sqrt((p+h)/p)
println("El valor de Q* con faltantes es=",Q2)
using Plots,DataFrames
pyplot()  
YQ=Array{Float64,1}(900)
OQ=Array{Float64,1}(900)
HQ=Array{Float64,1}(900)
AQ=Array{Float64,1}(900)
for i=1:900 
OQ[i]=i
YQ[i]=(i*h/2)+(A*D/i)+(c*D)
HQ[i]=h*i/2    
AQ[i]=(A*D)/i
end
print(DataFrame(Q =Q ,T=T,YQ=YK,K=K2))
plot(OQ,YQ,label="Y(Q)", line=(2),xlabel="Cantidad (Q)",ylabel="Costo")
plot!(OQ,HQ,label="hQ/2", line=(2))
plot!(OQ,AQ,label="AD/Q", line=(2))
plot!([894.42],label="Q*", line=(:dash,2),seriestype=:vline)
```

# Extensiones EOQ

# EPQ economic production quatity (EPL) o EPL (economic production lot (EPL))

La fórmula original de Harris se ha ampliado de diversas maneras a lo largo de los años. Una de las primeras extensiones (Taft 1918) fue en el caso en que la reposición no es instantánea; en cambio, existe una tasa de producción finita, pero constante y determinista. Este modelo a veces se denomina modelo de lote de producción económica (EPL). Si dejamos que P represente la tasa de producción (y asumimos que P> D para que el sistema tenga capacidad para satisfacer la demanda), el modelo EPL da como resultado el siguiente tamaño de lote para minimizar la suma de los costos de mantenimiento y pedido.

$Q^{*}=\sqrt{\frac{2AD}{h(1-D/P)}}$

Tenga en cuenta que si P = ∞ (es decir, la reposición es infinitamente rápida), entonces esta fórmula se reduce a la EOQ regular. De lo contrario, resulta en un tamaño de lote más grande para cubrir el hecho de que los artículos de reposición tardan en producirse.

# Modelos Dinamicos de tamaño de lote   

Como se señala anteriormente, la formulación de EOQ se basa en una serie de supuestos, específicamente,
1. Producción instantánea. 
2. Entrega inmediata.
3. La demanda determinista.
4. Demanda constante.
5. Costos constantes de preparacion de pedido.
6. Producto único o productos separables.

La introducción de retrasos en la entrega es sencilla si los tiempos de entrega son conocidos y fijos (es decir, calcular las cantidades de pedido de acuerdo con la fórmula EOQ y realizar los pedidos en momentos iguales a la entrega deseada menos el tiempo de entrega). 

Si los tiempos de entrega son inciertos, entonces se requiere un enfoque diferente. 

Sin embargo, la demanda es una fuente de aleatoriedad más prevalente e importante que los tiempos de entrega. El tema de relajar el supuesto de demanda determinista se tratará en la siguiente sección sobre modelos de inventario estadístico. 

## Wagner-Whitin

El principal enfoque histórico para relajar el supuesto de demanda constante es el modelo de Wagner-Whitin (Wagner y Whitin 1958).

harvey wagner es conocido por sus contribuciones a la programacion dinamica en el campo academio, por su proposicion de "problema de la diligencia"

Este modelo considera el problema de determinar los tamaños de los lotes de producción cuando la demanda es determinista pero varía en el tiempo y todas las otras suposiciones para el modelo EOQ son válidas.

La importancia de este enfoque dinámico de tamaño de lote es que ha tenido un impacto sustancial en la literatura, y luego influyó en el desarrollo de la planificación de necesidades de materiales (MRP) y la planificación de recursos empresariales (ERP). 

### Formulacion del modelo wagner-whitin

Cuando la demanda varía con el tiempo, un modelo de tiempo continuo, como el modelo EOQ, es incómodo de especificar. Entonces, en lugar de eso, agruparemos la demanda en períodos discretos, que podrían corresponder a días, semanas o meses, según el sistema. 

Un programa de producción diario puede tener sentido para un sistema de alto volumen con una demanda que cambia rápidamente, mientras que un programa mensual puede ser adecuado para un sistema de bajo volumen con una demanda que cambia más lentamente. 


Para especificar el problema y el modelo, utilizaremos la siguiente notación, que representa la contraparte dinámica de la notación estática utilizada para el modelo EOQ:

$t= $ un período de tiempo (asumiremos una semana, pero se podría usar cualquier intervalo); el rango de períodos de tiempo es t = 1, ..., T, donde T representa el horizonte de planificación

$D = $ Demanda en la semana t (en unidades)

$ct = $ costo unitario de producción (en dinero por unidad), sin contar los costos de solicitar pedido o costo de inventario en la semana t

$At = $ costo de solicitar (pedido) costo para producir (comprar) un lote en la semana t (en dinero) 

$ht = $ costo de mantenimiento para llevar una unidad de inventario de la semana t a la semana t + 1 (en dinero por unidad por semana); por ejemplo, si el costo de tenencia consiste completamente en intereses sobre dinero atado al inventario, donde i es la tasa de interés anual, entonces ht = i*ct / 52

$It = $ inventario (en unidades) sobrantes al final de la semana t

$Qt = $ tamaño del lote (en unidades) en la semana t; Hay T tales variables de decisión, una para cada semana.
