```julia
function costoc(A)
###Agregamos la informacion
tareas=[1; 2; 3; 4]
tiempo=[10; 8; 6 ;7]
fecha=[15; 20; 10; 30]
costoalmacen=[3; 2; 5; 4]
costopenalizacion=[10; 22; 10; 8]
###Imprimimos en una tabla para visualizar la informacion
M=[tareas tiempo fecha costoalmacen costopenalizacion]
#print(DataFrame(M))
#####Declaramos la solucion inicial

t=tiempo[A[1]]
fe=fecha[A[1]]
c1=fe-t
###programamos el costo para el primer trabajo en almacen y penalizacion
if c1>=0
    costo1=c1*costoalmacen[A[1]]
elseif c1==0
    costo1=0
else
    costo1=c1*(-costopenalizacion[A[1]])
end
#    println("costo P1= ",costo1)

###programamos el costo para el segundo trabajo en almacen y penalizacion

t2=tiempo[A[2]]+tiempo[A[1]]
fe2=fecha[A[2]]
c2=fe2-t2

if c2>=0
    costo2=c2*costoalmacen[A[2]]
elseif c2==0
    costo2=0
else
    costo2=c2*(-costopenalizacion[A[2]])
end
   # println("costo P2= ",costo2)

###programamos el costo para el tercer trabajo en almacen y penalizacion

t3=tiempo[A[3]]+tiempo[A[2]]+tiempo[A[1]]
fe3=fecha[A[3]]
c3=fe3-t3

if c3>=0
    costo3=c3*costoalmacen[A[3]]
elseif c3==0
    costo3=0
else
    costo3=c3*(-costopenalizacion[A[3]])
end
    #println("costo P3= ",costo3)

###programamos el costo para el cuarto trabajo en almacen y penalizacion

t4=tiempo[A[4]]+tiempo[A[3]]+tiempo[A[2]]+tiempo[A[1]]
fe4=fecha[A[4]]
c4=fe4-t4

if c4>=0
    costo4=c4*costoalmacen[A[4]]
elseif c4==0
    costo4=0
else
    costo4=c4*(-costopenalizacion[A[4]])
end
    #println("costo P4= ",costo4)

#el costo final de la primera iteracion es

ctot=costo1+costo2+costo3+costo4
    #println("costo total es= ",ctot)
# println(A)
end

function vecindades(S)
S1=[S[2] S[1] S[3] S[4]]
S2=[S[1] S[3] S[2] S[4]]
S3=[S[1] S[2] S[4] S[3]]
vecindades=[S1; S2; S3 ]
    return(vecindades)
end

function decision(vecindades,r)  
    if 1/3 >= r >= 0
  Sv=vecindades[1,:]
    elseif 2/3 >= r >= 1/3
    Sv=vecindades[2,:]
    else 1>= r >= 2/3
     Sv=vecindades[3,:]
    end
    return(Sv)
end
```
